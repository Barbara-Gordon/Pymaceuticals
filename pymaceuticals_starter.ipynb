```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset

combined_data = pd.merge(clinical_data, mouse_data, on = 'Mouse ID', how = 'left')

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

drug_time = combined_data[['Drug', 'Timepoint','Tumor Volume (mm3)']]
drug_time = drug_time.groupby(['Drug','Timepoint']).aggregate({'Tumor Volume (mm3)':'mean'}) 




# Convert to DataFrame
drug_time.head()
# Preview DataFrame
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_errorsDT = combined_data[['Drug', 'Timepoint','Tumor Volume (mm3)']]
standard_errorsDT = standard_errorsDT.groupby(['Drug','Timepoint']).aggregate({'Tumor Volume (mm3)': sem}) 
# Convert to DataFrame
standard_errorsDT.head()

# Convert to DataFrame

# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_pivot = pd.pivot_table(drug_time, values='Tumor Volume (mm3)', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked

standard_errorsDT_pivot = pd.pivot_table(standard_errorsDT, values='Tumor Volume (mm3)', index=['Timepoint'], columns=['Drug'])


drug_pivot1 = drug_pivot[['Capomulin' ,'Infubinol', 'Ketapril', 'Placebo']]
standard_errorsDT_pivot1 = standard_errorsDT_pivot[['Capomulin' ,'Infubinol', 'Ketapril', 'Placebo']]
drug_pivot1.head()
```

```python
standard_errorsDT_pivot1
```

```python
# Generate the Plot (with Error Bars)

drug_timeplot = drug_pivot1.plot(kind = 'line', yerr = standard_errorsDT_pivot1,figsize = (10.0, 7.0))

# Save the Figure

plt.grid()
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')
plt.legend(loc = 'best')


plt.savefig("../Pymaceuticals/TumorVolume.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 


drug_timeMS = combined_data[['Drug', 'Timepoint','Metastatic Sites']]
drug_timeMS = drug_timeMS.groupby(['Drug','Timepoint']).aggregate({'Metastatic Sites':'mean'}) 

# Convert to DataFrame

# Preview DataFrame
drug_timeMS.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

standard_errorsMS = combined_data[['Drug', 'Timepoint','Metastatic Sites']]
standard_errorsMS = standard_errorsMS.groupby(['Drug','Timepoint']).aggregate({'Metastatic Sites': sem}) 
# Convert to DataFrame
standard_errorsMS.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_pivotMS = pd.pivot_table(drug_timeMS, values='Metastatic Sites', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked
drug_pivotMS1 = drug_pivotMS[['Capomulin' ,'Infubinol', 'Ketapril', 'Placebo']]
standard_errorsMS_pivot = pd.pivot_table(standard_errorsMS, values='Metastatic Sites', index=['Timepoint'], columns=['Drug'])
standard_errorsMS_pivot1 = standard_errorsMS_pivot[['Capomulin' ,'Infubinol', 'Ketapril', 'Placebo']]
```

```python
drug_pivotMS1
```

```python
# Generate the Plot (with Error Bars)

drug_timeplotMS = drug_pivotMS1.plot(kind = 'line', yerr = standard_errorsMS_pivot1, figsize = (10.0, 7.0))

# Save the Figure

plt.grid()
plt.xlabel('Timepoint')
plt.ylabel('Metastatic Sites')
plt.title('Metastatic Sites Over Time')
plt.legend(loc = 'best')


plt.savefig("../Pymaceuticals/MetastaticSites.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

survival_rate = combined_data[['Drug', 'Timepoint', 'Mouse ID']]
survival_rate = survival_rate.groupby(['Drug','Timepoint']).aggregate({'Mouse ID':'count'}) 
survival_rate = survival_rate.rename(columns = {'Mouse ID':'Mouse Count'})
# Convert to DataFrame

# Preview DataFrame

survival_rate.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

survival_ratepivot = pd.pivot_table(survival_rate, values='Mouse Count', index=['Timepoint'], columns=['Drug'])
# Preview that Reformatting worked
survival_ratepivot1 = survival_ratepivot[['Capomulin' ,'Infubinol', 'Ketapril', 'Placebo']]

survival_ratepivot1
```

```python
survival_ratepivot
```

```python
# Generate the Plot (with Error Bars)


survival_rate_plot = survival_ratepivot1.plot(kind = 'line', figsize = (10.0, 7.0))

# Save the Figure

plt.grid()
plt.xlabel('Timepoint')
plt.ylabel('Mouse Count')
plt.title('Survival Rate by Drug')
plt.legend(loc = 'best')


plt.savefig("../Pymaceuticals/SurvivalRates.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

summary_data = ((drug_pivot.loc[45,:] - drug_pivot.loc[0,:])/drug_pivot.loc[0,:])*100
print(summary_data)

#survival_rate_percentage = (survival_rate_data - survival_rate_data['Placebo'])/ survival_rate_data['Placebo']
#print(survival_rate_percentage)
# Display the data to confirm
```

```python
# Store all Relevant Percent Changes into a Tuple

percent_data = summary_data.items()

values_listT= []

for v in percent_data:
    values_listT.append(v) 


labels, values = zip(*values_listT)

colors = []

for v in values:
    if v < 0:
        colors.append('palegreen')
    else:
        colors.append('red')
        
x_axis = np.arange(0,len(labels),1)

ax = summary_data.plot(kind = 'bar', color = colors, figsize = (25.0, 7.0), grid = True, width = .95)
ax.set_title('Change in Tumor Size (as Percentage)',fontsize = 30)
ax.set_xlim(-.75,9.75)
ax.set_ylim(-25,65)
ax.set_xlabel('Drug', fontsize = 20)
ax.set_ylabel('Percentage Change',fontsize = 20)
ax.set_xticklabels(labels,fontsize = 16, rotation = 'horizontal')
 

i = -0.15

for v in values:
    vf = format(v, ".2f") + '%'
    ax.text(i,(v+.85),vf,fontsize = 16)
    i +=1
    
# Save the Figure

plt.savefig("../Pymaceuticals/SummaryGraph.png")

# Show the Figure
```
